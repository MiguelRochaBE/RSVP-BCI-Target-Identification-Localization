```python
from scipy.io import loadmat
import numpy as np

# Load the .mat files
tERP_seatedCz = loadmat('tERP_seatedCz.mat')
ntERP_seatedCz = loadmat('ntERP_seatedCz.mat')

tERP_seatedPO8 = loadmat('tERP_seatedPO8.mat')
ntERP_seatedPO8 = loadmat('ntERP_seatedPO8.mat')

# Access the variables
tERP_seatedCz  = tERP_seatedCz['lol'] # lol is the variable name that they were saved as in Matlab
ntERP_seatedCz  = ntERP_seatedCz['lol']

tERP_seatedPO8  = tERP_seatedPO8['lol']
ntERP_seatedPO8  = ntERP_seatedPO8['lol']
```

## **Central**

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

tex = np.linspace(0, len(tERP_seatedCz), len(tERP_seatedCz))

figure(figsize=(12, 4), dpi = 600);
plt.plot(tex*1000/125-200, tERP_seatedCz, color = 'r', label = "Target");
plt.plot(tex*1000/125-200, ntERP_seatedCz, color = 'b', label = "Non-target");
plt.xlabel("Time (ms)", fontsize = 12);
plt.ylabel("Amplitude ($\mu V$)", fontsize = 12);
plt.title("Cz", fontsize = 12);
plt.xticks(np.arange(-200,801, step = 100),fontsize = 12);
plt.yticks(fontsize = 12);
plt.axhline(0, color = 'k', linestyle = '--');
plt.axvline(0, color = 'k', linestyle = '--');

# Define the coordinates and sizes of the rectangles
rectangles = [
    # First tuple: (x,y) | Second tuple: ( width and height)
    plt.Rectangle((90, -10), 45, 20, color='green', alpha=0.2, label = "VPP"), 
    plt.Rectangle((180, -10), 50, 20, color='blue', alpha=0.2, label = "N2"),
    plt.Rectangle((250, -10), 100, 20, color='orange', alpha=0.2, label = "P3")
    ]

# Add the rectangles to the plot
for rect in rectangles:
    plt.gca().add_patch(rect)

plt.legend(fontsize = 11)
plt.show()
```
